From the provided result:We Observed

5. **Disposable Income Variation**:

Overall, these observations provide valuable insights into the income and expenditure patterns of the sample population, which can be further analyzed to understand consumer behavior and inform strategic decisions related to retail sales growth.

```python
# Calculate total sample size, total income, total expenditure

total_sample_size = credpal['Number of Samples'].sum()
total_income = credpal['Total Income of Sample size ( $)'].sum()
total_expenditure = credpal['Total Expenditure of Sample Size (00 $)'].sum()
total_disposable_income = credpal['Disposable Income of Sample Size (00 $)'].sum()

print("Number of Samples:", total_sample_size)
print("Total Income of Sample size:", total_income)
print("Total Expenditure of Sample Size:", total_expenditure)
print("Total Disposable Income of Sample Size:", total_disposable_income)
```

```python
print(credpal.dtypes)
```

```python
credpal['Study Time: Start'] = pd.to_datetime(credpal['Study Time: Start'])
credpal['Study Time: Ends'] = pd.to_datetime(credpal['Study Time: Ends'])
```

```python
credpal.info()
```

```python
#Top 5 Total Disposable Income by Consumer Preference
disposable_income_by_consumer_preference = credpal.groupby('Consumer Preference')['Disposable Income of Sample Size (00 $)'].sum()
top_consumer_preference = disposable_income_by_consumer_preference.sort_values(ascending=False).head(5)
print("Top 5 Total Disposable Income by Consumer Preference:")
print(top_consumer_preference)
```

```python
#Top 5 Countries by expenditure
expenditure_by_countries = credpal.groupby('Country')['Total Expenditure of Sample Size (00 $)'].sum()
top_countries_by_expenditure = expenditure_by_countries.sort_values(ascending=False).head(5)
print("Total expenditure by country:")
print(top_countries_by_expenditure)
```

```python
credpal.info()
```

```python
#Top 5 disposable income by region
disposable_income_by_region = credpal.groupby('Region')['Disposable Income of Sample Size (00 $)'].sum()
top_region_by_disposable_income = disposable_income_by_region.sort_values(ascending=False).head(5)
print("Total disposable Income by region:")
print(top_region_by_disposable_income)
```

```python
#Total Income by region
total_income_by_region = credpal.groupby('Region')['Total Income of Sample size ( $)'].sum()
total_income_by_region_sorted = total_income_by_region.sort_values(ascending=True)
plt.figure(figsize=(9, 6))
total_income_by_region_sorted.plot(kind='barh', color='skyblue')
plt.xlabel('Region')
plt.ylabel('Total Income of Sample size ( $)')
plt.title('Total Income by Region')

for i, v in enumerate(total_income_by_region_sorted):
    plt.text(v, i, f'{v:,.2f}', ha='left', va='center', color='black')
    
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
#Average monthly income by consumer preference
total_average_monthly_income = credpal.groupby('Consumer Preference')['Average monthly income( $)'].sum()
total_average_monthly_income_sorted = total_average_monthly_income.sort_values(ascending=True)
plt.figure(figsize=(9, 6))
total_average_monthly_income_sorted.plot(kind='barh', color='skyblue')
plt.xlabel('Consumer Preference')
plt.ylabel('Average monthly income( $)')
plt.title('Average Monthly Income by Consumer Preference')

for i, v in enumerate(total_average_monthly_income_sorted):
    plt.text(v, i, f'{v:,.2f}', ha='left', va='center', color='black')

plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python

total_income_by_region = credpal.groupby('Region')['Total Income of Sample size ( $)'].sum()
total_expenditure_by_region = credpal.groupby('Region')['Total Expenditure of Sample Size (00 $)'].sum()

regions = total_income_by_region.index
num_regions = len(regions)

bar_width = 0.35


index = np.arange(num_regions)

plt.figure(figsize=(10, 6))


plt.bar(index, total_income_by_region, width=bar_width, color='skyblue', label='Total Income')


plt.bar(index + bar_width, total_expenditure_by_region, width=bar_width, color='blue', label='Total Expenditure')

# Set x-axis ticks and labels
plt.xlabel('Region')
plt.ylabel('Amount ($)')
plt.title('Total Income and Total Expenditure by Region')
plt.xticks(index + bar_width / 2, regions, rotation=45)  # Position ticks at the center of each group
plt.legend()
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

# INSIGHT
From the income and expenditure analysis, it is evident that consumer spending trends remain consistent across regions and countries, regardless of income size, indicating scalability potential for CredPal's credit solutions worldwide.

Sub-Saharan Africa stands out with the highest income aggregate, correlating with increased saving capability. This suggests that higher disposable incomes are associated with greater savings potential.

Notably, office supplies and household products dominate total expenditure, indicating a preference for investments in business and housing over consumables. Consequently, expanding credit solutions across regions promises profitability for CredPal.
